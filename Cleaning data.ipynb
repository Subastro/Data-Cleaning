```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df= pd.read_excel("Downloads/query 2.xlsx")
df.head()
```

```python
df.info()
```

```python
#drop the columns "nst" since it doesnot have any data
df.drop(columns=["nst"],inplace=True)

df.info()
```

```python
#Drop all rows that contain missing values (NaN) from any column
df.dropna(inplace=True)
df.info()
```

```python
# Drop the 'status' column from the DataFrame
df.drop(columns=["status"],inplace=True)

# Drop the 'locationSource' column 
df.drop(columns=["locationSource"],inplace=True)

# Drop the 'magSource' column 
df.drop(columns=["magSource"],inplace=True)

# Drop the 'net' column 
df.drop(columns=["net"],inplace=True)
df.info()
```
